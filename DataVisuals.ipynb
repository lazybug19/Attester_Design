```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataFrame = pd.read_csv('/content/IoT_Dataset.csv')
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=dataFrame, x='Attack_label', hue='Attack_type', edgecolor='black', linewidth=1)
plt.title('Count of Attack Types vs Benign', fontsize=20)
plt.show()
```

Dataset has 0 variance; skipping density estimate

```python
counts = dataFrame.groupby(['Attack_label', 'Attack_type']).size().reset_index(name='counts')
plt.figure(figsize=(15, 6))
sns.kdeplot(data=counts, x='counts', hue='Attack_type', fill=True, common_norm=False, palette="crest", alpha=.5, linewidth=2)
plt.title('KDE Plot of Attack Types vs Benign', fontsize=20)
plt.xlabel('Count')
plt.ylabel('Density')
plt.show()
```

```python
counts = dataFrame.groupby(['Attack_label', 'Attack_type']).size().reset_index(name='counts')
plt.figure(figsize=(15, 6))
sns.boxplot(data=counts, x='Attack_label', y='counts', hue='Attack_type', width=1)
plt.ylim(0, max(counts['counts']) + 20)
plt.title('Box Plot of Attack Types vs Benign', fontsize=20)
plt.xlabel('Attack Label')
plt.ylabel('Count')
plt.show()
```

```python
counts = dataFrame.groupby(['Attack_label', 'Attack_type']).size().reset_index(name='counts')

plt.figure(figsize=(15, 6))
sns.histplot(data=counts, x='counts', hue='Attack_type', multiple='stack', kde=False, palette="crest", edgecolor='black')
plt.xlabel('Count of Attack Type Classes and Normal Classes')
plt.ylabel('Frequency of Counts')
plt.show()
```
